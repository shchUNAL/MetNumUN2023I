## SymPy’s polynomials 
- https://docs.sympy.org/latest/modules/polys/basics.html#polynomials 

$$ P_n(x) =  a_n x^n + a_{n-1} x^{n-1} + a_{n-2} x^{n-2} +\cdots +  a_1 x  + a_0$$

$$P_n(x) = 0 $$

$$a_n x^n + a_{n-1} x^{n-1} + a_{n-2} x^{n-2} +\cdots +  a_1 x  + a_0 = 0$$

$$(x-r_1) Q_{n-1}(x) = P_n(x)$$
$$(x-r_1) (x - r_2) Q_{n-2}(x) = P_n(x)$$

**My own example 12**

p = 8p1+Ip2, q = 2q1+Iq2, r = -2r1+8Ir2, p/(qr)

monda monda monda monda monda monda monda.png

```
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)  
x = x1 + I*x2
y = y1 + I*y2
z = z1 + I*z2

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```

**My own example 13**

```
from sympy import symbols, I, N, pi, expand
p1, p2,q1,q2,r1,r2 = symbols("p1, p2,q1,q2,r1,r2", real=True)  
p=8*p1+I*p2
q=2*q1+I*q2
r=-2*r1+8*I*r2

print(p/q*r)
print(expand(p/q*r))
print(expand((p/q)*r))
print(expand(p/(q*r)))
x=1/(2*pi + I)
print('x=',x)
x=N(x)
print('x=',x)
```
